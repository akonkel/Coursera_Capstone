This notebook will use Beautiful Soup to read the table of Toronto zip codes from the wikipedia page listed in the assignment.  Let's start by importing the stuff we'll need.

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

Now let's get that table.

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
file = requests.get(url)
soup = BeautifulSoup(file.text)
for tag in soup.find_all(True):
    print(tag.name)
```

Looking at the printout, it looks like we want the table with class 'wikitable'.

```python
zips = soup.table
for string in zips.stripped_strings:
    print(repr(string))
```

Looks like the right stuff.  Now, we need to take apart the HTML and put it in the pandas dataframe as requested.  The data is in triplets with the first entry being the postal code, the second the borough, and the third the neighborhood(s).

```python
df = pd.read_html(str(zips))[0]
df
```

Lovely.  Now we want to clear out the entries with borough 'not assigned'.  I would also make the neigborhood the same as the borough if 'not assigned' and make sure there aren't any duplicate boroughs, but those don't apply (i.e., I checked and they don't happen so they don't need to be accounted for).

```python
df = df[df.Borough!='Not assigned']
df.reset_index(drop=True)
```
