```python
# part 1 code
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
file = requests.get(url)
soup = bs(file.text)
zips = soup.table
df = pd.read_html(str(zips))[0]
df = df[df.Borough!='Not assigned']
df.reset_index(drop=True)
```

```python
# part 2 code
# !pip install geocoder  #uncomment if you need to install
import geocoder # import geocoder


# make variables to hold lats and longs
latitude = []
longitude = []

# loop through rows of zip code data frame
for code in df['Postal code']:
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
      g = geocoder.arcgis('{}, Toronto, Ontario'.format(code))
      lat_lng_coords = g.latlng
      latitude.append(lat_lng_coords[0])
      longitude.append(lat_lng_coords[1])
        
df['Latitude'] = latitude
df['Longitude'] = longitude
df
```

Now I have a dataframe with postal codes, boroughs, neighborhoods, and the lat/longs for the postal code.  As suggested by the assignment, I'll limit to the boroughs that contain 'Toronto' and then cluster like we did in the lab assignment.

```python
# reduce data frame to just Toronto neighborhoods and map it
Toronto = df[df['Borough'].str.contains("Toronto")]

# we have lat/long sets from the previous loop for adding them to the zip list, so just use the last ones to start the map

map_toronto = folium.Map(location=[latitude[0], longitude[0]], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(Toronto['Latitude'], Toronto['Longitude'], Toronto['Borough'], Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
